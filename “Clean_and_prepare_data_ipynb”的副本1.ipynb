**Feature engineering**

```python
df = pd.read_csv('Joined_Dataframe_remove_duplicates.csv')
df.head()

# Feature engineering

invoice_agg = df.groupby('Invoice').agg({
    'Quantity': 'sum',
    'TotalAmount': 'sum',
    'StockCode': 'nunique',
    'Canceled': 'max',
    'Country': 'first'
}).reset_index()

# Rename columns for clarity.
invoice_agg.rename(columns={
    'Quantity': 'TotalQuantity',
    'StockCode': 'NumItems'
}, inplace=True)

# Create an average unit price column (handling division by zero).
invoice_agg['AvgUnitPrice'] = invoice_agg.apply(
    lambda row: row['TotalAmount'] / row['TotalQuantity'] if row['TotalQuantity'] != 0 else 0, axis=1)

print("\nAggregated (invoice-level) data shape:", invoice_agg.shape)
print(invoice_agg.head())

# For our classification, select only the numeric features and the target.
data_model = invoice_agg[['TotalQuantity', 'TotalAmount', 'NumItems', 'AvgUnitPrice', 'Canceled']]

print("\nData description for modeling:")
print(data_model.describe())

# #orders_feature
df['delivery_time'] = (pd.to_datetime(df['order_delivered_customer_date']) - pd.to_datetime(df['order_purchase_timestamp'])).dt.days
df['is_late_delivery'] = (pd.to_datetime(df['order_delivered_customer_date']) > pd.to_datetime(df['order_estimated_delivery_date'])).astype(int)
# Customer Characteristics
df['customer_order_count'] = df.groupby('customer_id')['order_id'].transform('count')
df['customer_avg_review_score'] = df.groupby('customer_id')['review_score'].transform('mean')
# Payment way
df['payment_method_count'] = df.groupby('customer_id')['payment_value'].transform('count')
df['payment_method_avg_value'] = df.groupby('customer_id')['payment_value'].transform('mean')
```
