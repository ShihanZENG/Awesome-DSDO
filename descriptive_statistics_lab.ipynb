## Introduction to Descriptive Statistics

Descriptive statistics provide a concise summary of the main characteristics of data, often with simple numbers. They offer insights into the data's main features, giving an overview that can help understand its meaning and context.


```python
import numpy as np
from scipy import stats

# Sample data
data = [2, 3, 4, 5, 5, 6, 7, 8, 9, 5, 3, 10, 11, 12, 15, 16, 14, 16, 17, 18, 20, 21, 22, 23, 24, 25, 26]
data
```

Frequency refers to the number of times a particular data value occurs. A frequency distribution organizes the dataset, indicating the count of each value or range of values.

## Measures of Central Tendency

Measures of central tendency describe the center of a dataset. The primary measures are:

1. **Mean (Arithmetic Average)**:
   - Formula:    - It represents the sum of all the data values divided by the number of data values.
   
2. **Median**:
   - It is the middle value when the data is sorted. For an even number of points, it's the average of the two middle numbers.
   
3. **Mode**:
   - The value that appears most frequently in the dataset.

These measures help identify a "typical" value in the dataset.


```python
# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Mode
mode = stats.mode(data)

mean, median, mode
```

```python
# Create a descriptive statistics summary
stats_summary = {
    'count': len(data),
    'mean': np.mean(data),
    'std': np.std(data, ddof=1), # ddof=1 provides sample standard deviation
    'min': np.min(data),
    '25%': np.percentile(data, 25),
    'median': np.median(data),
    '75%': np.percentile(data, 75),
    'max': np.max(data),
    'var': np.var(data, ddof=1), # ddof=1 provides sample variance
    'range': np.ptp(data), # ptp stands for peak to peak (max - min)
    'iqr': stats.iqr(data), # Interquartile range (Q3 - Q1)
    'skewness': stats.skew(data),
}

stats_summary
```


## Measures of Dispersion

Measures of dispersion provide insights into the spread or variability of data. They include:

1. **Range**:
   - Formula: Range(x)=Max(x)-Min(x)
   - Represents the difference between the highest and lowest values.
   
2. **Variance**:
   - Formula for a sample:

   image.png
   - Measures the average of the squared differences from the mean.
   
3. **Standard Deviation**:
   - Formula for a sample:

   image.png
   - Provides a measure of the average distance each data point is from the mean.

These measures help understand how spread out the values are in the dataset.


```python
# Range
range_val = np.ptp(data)

# Variance
variance = np.var(data, ddof=1)

# Standard Deviation
std_dev = np.std(data, ddof=1)

range_val, variance, std_dev
```


## Interquartile Range (IQR)

IQR measures the statistical spread between the 25th percentile (Q1) and the 75th percentile (Q3) in a dataset. It gives insights into the middle 50% of the dataset and is robust against outliers.

Formula:
IQR = Q3 - Q1


```python
# IQR calculation
Q3 = np.percentile(data, 75)
Q1 = np.percentile(data, 25)
IQR = Q3 - Q1
IQR
```


## Visualization of Data

Visualizing data helps in understanding its distribution and identifying patterns, outliers, or anomalies. Common visualizations include:

- **Histogram**: Represents the frequency distribution of continuous variables.
- **Boxplot**: Showcases data spread and spots outliers.

Let's visualize our sample data with these plots.


image.png

```python
import matplotlib.pyplot as plt
# Histogram
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(data, bins='auto', color='blue', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
# Boxplot
plt.subplot(1, 2, 2)
plt.boxplot(data, vert=False)
plt.title('Boxplot')
plt.tight_layout()
plt.show()
```

# 1. Arithmetic Mean
**Problem:** A company operating in the automotive industry has generated the following revenues last year. The government is planning to extra tax the companies that had an average monthly revenue more than £40.000. Should this company apply for the financial aid package?



image.png	image.png

**Decision:** Apply for the financial aid package

One can now conclude that the average monthly revenue for the company is £32,888. The company can apply for the financial aid package.


```python
import matplotlib.pyplot as plt
import statistics

# Monthly revenue data in GBP (£)
revenues = {
    "January": 26010.71,
    "February": 14794.74,
    "March": 52045.05,
    "April": 26254.18,
    "May": 52890.76,
    "June": 29549.54,
    "July": 42755.64,
    "August": 27376.99,
    "September": 34836.61,
    "October": 36249.06,
    "November": 38691.00,
    "December": 13200.00
}

# Calculate the average monthly revenue
average_revenue = sum(revenues.values()) / len(revenues) #alternatively: average_revenue = statistics.mean(revenues.values())

# Determine if the company should apply for the financial aid package
should_apply_for_aid = average_revenue <= 40000

# Create a bar chart for the monthly revenues
months = list(revenues.keys())
monthly_revenues = list(revenues.values())

plt.figure(figsize=(12, 6))
plt.bar(months, monthly_revenues, color='skyblue')
plt.axhline(average_revenue, color='red', linestyle='--', label=f'Average Revenue (£{average_revenue:.2f})')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.title('Monthly Revenues for the Year 2022')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()

# Output decision on financial aid application
decision = 'Yes' if should_apply_for_aid else 'No'
print(f"The average monthly revenue is: £{average_revenue:.2f}")
print(f"Should the company apply for the financial aid package? {decision}")
```

image.png

image.png

**Issue:**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Monthly revenue data in GBP (£)
monthly_data = {
    "Month": [
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ],
    "Revenue": [
        26010.71, 14794.74, 52045.05, 26254.18, 52890.76, 29549.54,
        42755.64, 27376.99, 34836.61, 36249.06, 10106.00, 140000.00
    ]
}

# Convert to DataFrame
df = pd.DataFrame(monthly_data)

# Calculate the average monthly revenue
average_revenue = df['Revenue'].mean()

# Determine if the company should apply for the financial aid package
should_apply_for_aid = average_revenue <= 40000

# Create a bar chart for the monthly revenues
plt.figure(figsize=(12, 8))
plt.bar(df['Month'], df['Revenue'], color='skyblue')
plt.axhline(average_revenue, color='red', linestyle='--', label=f'Average Revenue (£{average_revenue:.2f})')
plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.title('Monthly Revenues for the Year 2022')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Output decision on financial aid application
decision = 'Yes' if should_apply_for_aid else 'No'
average_revenue, decision
```

# 6. Coefficient of variation (Standard deviation)
**Problem:** Two companies are evaluated based on their revenue performance. Which one seems to be doing better?

image.png!
image.png

**Decision:** Company B is more consistent

While company B has a higher average revenue, the process is not as consistent. The revenues are more spread, thus, it is more difficult to predict how the revenue generation process will behave in the future.


```python
import pandas as pd

# Revenue data for Company A and Company B
months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
company_a_revenue = [
    24653, 19670, 22659, 24566, 23849, 18738,
    18864, 21876, 19687, 18569, 17541, 22567
]
company_b_revenue = [
    26063, 38569, 24357, 28694, 22332, 11059,
    29756, 19659, 36591, 27421, 20212, 11341
]

# Create a DataFrame
df = pd.DataFrame({
    'Month': months,
    'Company A': company_a_revenue,
    'Company B': company_b_revenue
})

# Calculate the mean and standard deviation for each company
mean_a = df['Company A'].mean()
std_a = df['Company A'].std()
mean_b = df['Company B'].mean()
std_b = df['Company B'].std()

# Calculate the Coefficient of Variation for each company
cv_a = std_a / mean_a
cv_b = std_b / mean_b

# Generate a bar chart for monthly revenue
plt.figure(figsize=(14, 7))
bar_width = 0.35
index = np.arange(len(months))
bar1 = plt.bar(index, company_a_revenue, bar_width, label='Company A', color='b')
bar2 = plt.bar(index + bar_width, company_b_revenue, bar_width, label='Company B', color='r')

plt.xlabel('Month')
plt.ylabel('Revenue (£)')
plt.title('Monthly Revenue of Company A vs Company B')
plt.xticks(index + bar_width / 2, months, rotation=45)
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

cv_a, cv_b
```
